1. Create a random 4x4 array from an exponential distribution. Refer numpy functions.

```python
import numpy as np
# Create a random 4x4 array from an exponential distribution
# using numpy.random.exponential()
exponential_array = np.random.exponential(scale=1.0, size=(4, 4))
print("Random 4x4 array from exponential distribution:")
print(exponential_array)
```

Explanation:

np.random.exponential(scale=1.0, size=(4, 4)):
np.random: This accesses the random number generation module within NumPy.
exponential(): This function generates random samples from an exponential distribution.
scale=1.0: This parameter determines the scale parameter (beta) of the exponential distribution. The default is 1.0.
size=(4, 4): This parameter specifies the shape of the output array. We want a 4x4 matrix, so we set size=(4, 4).
The generated random array is stored in the variable exponential_array.
print(...): These lines simply print a descriptive message and the resulting exponential_array.

2. Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Create a random 100000x1 array from the same exponential distribution
exponential_data = np.random.exponential(scale=1.0, size=100000)

# plt.hist(np.random.rand(100000), density=True, bins=100, histtype="step", color="blue", label="Uniform")
# plt.hist(np.random.randn(100000), density=True, bins=100, histtype="step", color="red", label="Normal")
plt.hist(exponential_data, density=True, bins=100, histtype="step", color="green", label="Exponential")


plt.axis([0, 10, 0, 1.1]) # Adjust the x-axis range to better visualize the exponential distribution
plt.legend(loc = "upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```
