1. Create a random 4x4 array from an exponential distribution. Refer numpy functions.

```python
import numpy as np
# Create a random 4x4 array from an exponential distribution
# using numpy.random.exponential()
exponential_array = np.random.exponential(scale=1.0, size=(4, 4))
print("Random 4x4 array from exponential distribution:")
print(exponential_array)
```

Explanation:

np.random.exponential(scale=1.0, size=(4, 4)):
np.random: This accesses the random number generation module within NumPy.
exponential(): This function generates random samples from an exponential distribution.
scale=1.0: This parameter determines the scale parameter (beta) of the exponential distribution. The default is 1.0.
size=(4, 4): This parameter specifies the shape of the output array. We want a 4x4 matrix, so we set size=(4, 4).
The generated random array is stored in the variable exponential_array.
print(...): These lines simply print a descriptive message and the resulting exponential_array.

  3. Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5].

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a grid of x and y values
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

# Calculate Z
Z = X**2 + Y**2

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D plot of Z = X^2 + Y^2')

plt.show()
```

**Explanation:**

*   **`import matplotlib.pyplot as plt` and `import numpy as np`**: These lines import the necessary libraries: `matplotlib.pyplot` for plotting and `numpy` for numerical operations.
*   **`x = np.linspace(-5, 5, 100)` and `y = np.linspace(-5, 5, 100)`**: These create arrays of 100 evenly spaced values between -5 and 5 for both the x and y axes.
*   **`X, Y = np.meshgrid(x, y)`**: This creates a grid from the x and y arrays. `X` and `Y` are 2D arrays where each element corresponds to a point on the grid.
*   **`Z = X**2 + Y**2`**: This calculates the corresponding Z values for each (X, Y) point on the grid based on the equation Z = X² + Y².
*   **`fig = plt.figure()`**: This creates a new figure for the plot.
*   **`ax = fig.add_subplot(111, projection='3d')`**: This adds a 3D subplot to the figure.
*   **`ax.plot_surface(X, Y, Z, cmap='viridis')`**: This plots the 3D surface using the calculated X, Y, and Z values. `cmap='viridis'` sets the color map for the surface.
*   **`ax.set_xlabel('X')`, `ax.set_ylabel('Y')`, `ax.set_zlabel('Z')`, `ax.set_title('3D plot of Z = X^2 + Y^2')`**: These lines set the labels for the axes and the title of the plot.
*   **`plt.show()`**: This displays the generated plot.
